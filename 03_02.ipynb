```python
# Start by importing our libraries
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# Now let's build a quick plot
diamonds = pd.read_csv('data/diamonds.csv')
fig = px.histogram(diamonds, x='price')
fig.show()
```

```python
# Create a Facets (trellis) for our chart
fig = px.histogram(diamonds, x='price', facet_row='clarity', facet_col='cut')
fig.show()
```
